## Отбираем самые скоррелированные признаки и посмотрим матрицу корреляции по каждому источнику


```python
def data_processing(data,target,min_corr,source):
    """
    Функция соеденяет фичи с целевой переменной, и отбирает признаки, с
    которые сильнее всех коррелируют с целевой переменной. И записывает результат в
    дирректорию \data_processing\

    Arguments:
      data: данные из источника (фичи) DataFrame 
      target: целевая переменная DataFrame
      min_corr: минимальный коэффициент корреляции flaut
      source: название источника str
    
    Returns:
      Pandas DataFrame

    """
    y = target['new_cases']
    data['y']=y
    corr_y = data.corr()[data.shape[1]-1:].T
    corr_y=corr_y.sort_values(by='y',ascending=False)
    corr_y.loc[corr_y['y'] > min_corr]
    list=corr_y.loc[corr_y['y'] > min_corr].index.to_list()
    source_d=data[list]
    list_new = ['{}_{}'.format(source,number) for number in list]
    source_d= source_d.rename(columns=dict(zip(list, list_new)))
    source_d.to_csv('processed_data\\{}_DE.csv'.format(source))
    corr_source_d = source_d.corr()
    return corr_source_d
```

## Google Trends

```python
data = pd.read_excel("raw_data\\google_de.xlsx",
                 index_col=0)
target = pd.read_excel("raw_data\\target_de.xlsx",  
               index_col=0)

target=target[:data.shape[0]]

min_corr=0.7
source='google'

data_processing(data,target,min_corr,source).style.background_gradient(cmap='coolwarm')
```

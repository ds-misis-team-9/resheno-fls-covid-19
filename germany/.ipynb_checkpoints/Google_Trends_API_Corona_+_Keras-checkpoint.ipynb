```python
from pytrends.request import TrendReq
import pandas as pd
import datetime as dt
import numpy as np
 
# скачаем данные с google doc файла
data = pd.read_csv('https://docs.google.com/spreadsheets/d/10WgVF3KSGS969coU-YMMjY2b_NIUEAPz0jFU1OApwPs/export?format=csv&gid=0')
data=data.loc[data['trends'] == 1]

# возьмём значения по ключевым словам-запросам
kw_list = data['Запрос'].values.tolist()
#kw_list = ['потеря обоняния', 'потеря обоняния при ковид' ]
pytrends = TrendReq(hl='RU', tz=180)
```

```python
# Функция вычисления тренда 

def trends (kw_list):
    pytrends.build_payload(kw_list, timeframe='2020-01-26 2021-04-25', geo='RU', gprop='')
    iot_df = pytrends.interest_over_time()
    iot_df = iot_df.reset_index()
    iot_df = iot_df.drop(['isPartial'], axis=1)
    return iot_df

# Функция сборки всего датасета (step - максимально разрешённое количество одновременных запросов, здесь - 5)

def step_data(step):
    query_df = trends(kw_list[:1])[['date']]
    p=step
    for i in range(0, len(kw_list), p):
        df = trends(kw_list[i:i+p])
        query_df = pd.concat([query_df, df.drop(['date'],axis=1)], axis=1)    
    return query_df
```

## 3. Применение различных моделей и их оценка точности (по MSE и RMSE)

```python
# разделим выборку на тренировочную и тестовую, и номализуем значения

from sklearn.model_selection import train_test_split

x = df.iloc[:,1:-1].values    # массив признаков

Y = df.iloc[:,-1].values      # массив значений

# нормализуем значения RobustScaler (по медиане т.к. могут быть выбросы) 

from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

#scaler = RobustScaler()
scaler = StandardScaler()
x_norm = scaler.fit_transform(x)
Y_norm = scaler.fit_transform(Y.reshape(-1,1))
X_norm = np.hstack([np.ones(x_norm.shape[0])[:, np.newaxis], x_norm])

# разбиение на выборки 80:20
X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y_norm, test_size = 0.2, random_state = 77) 

#y_train = np.ravel(Y_train, order='K')
#y_test = np.ravel(Y_test, order='K')

print(np.shape(X_train.round(2)), np.shape(X_test.round(2)))
#y_test, X_test
```

```python
from sklearn.metrics import mean_squared_error, f1_score, accuracy_score, roc_curve, roc_auc_score

# функция вывода значения метрик (в случае с регрессией - MSE и RMSE)

def print_regression_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f'MSE = {mse:.2f}, RMSE = {rmse:.2f}')
```

```python
# Линейная регрессия с оптимизацией theta градиентный спуском

# Функция вычисления градиента функции MSE

def calc_mse_gradient(X, y, theta):
    n = X.shape[0]
    grad = 1. / n * X.transpose().dot(X.dot(theta) - y)
    
    return grad

# Функция, осуществляющая градиентный шаг
# (параметр величины шага alpha - learning rate)

def gradient_step(theta, theta_grad, alpha):
    return theta - alpha * theta_grad

# Функция цикла градиентного спуска с доп. параметрами
# начального вектора theta и числа итераций

def optimize(X, y, grad_func, start_theta, alpha, n_iters):
    theta = start_theta.copy()
    
    for i in range(n_iters):
        theta_grad = grad_func(X, y, theta)
        theta = gradient_step(theta, theta_grad, alpha)
    
    return theta
```

```python
m = X_train.shape[1]
theta = optimize(X_train, y_train, calc_mse_gradient, np.ones(m), 0.001, 10000)
theta
```

```python
y_pred = X_test.dot(theta)
y_pred
```

```python
print_regression_metrics(y_test, y_pred)
```

```python
# Решающее дерево

from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(max_depth=4).fit(X_train,y_train)
y_train_predict = model.predict(X_test) 

print_regression_metrics(y_test, y_train_predict)
```

```python
# Линейная регрессия без оптимизации

def linreg_linear(X, y):
    theta = np.linalg.inv(X.T@X)@X.T@y
    return theta

theta2 = linreg_linear(X_train, y_train)

y_pred2 = X_test.dot(theta2)

print_regression_metrics(y_test, y_pred2)
```

```python
# Тестирование максимально возможной оценки на собственной выборке

y_train_predict2 = model.predict(X_train) 

print_regression_metrics(y_train, y_train_predict2)
```

```python
# ансамблевый метод - случайный лес с оптимизацией/подбором параметров

from sklearn.ensemble import RandomForestRegressor
from pprint import pprint
rf = RandomForestRegressor(random_state = 42)

print('Параметры п умолчанию:\n')
pprint(rf.get_params())
```

```python
# зададим сетку параметров, которые будут перебираться

from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
max_features = ['auto', 'sqrt', 'log2']
max_depth = [int(x) for x in np.linspace(4, 10, num = 1)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf = RandomForestRegressor()

rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, 
                               cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
rf_best = RandomForestRegressor(n_estimators = 50, 
                                min_samples_split = 2, 
                                min_samples_leaf= 2,
                                max_features= 'auto',
                                max_depth= 4,
                                bootstrap= False,
                                random_state=1,
                                criterion = 'mse')
rf_best.fit(X_train, y_train)
y_best_pred = rf_best.predict(X_test)
mse_best = mean_squared_error(y_test, y_best_pred)

print(mse_best)
print_regression_metrics(y_test, y_best_pred)
```

```python
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
import tensorflow as tf
import keras
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.backend import clear_session
import sklearn
import matplotlib.pyplot as plt
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
clear_session()
```

```python
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],))) # первый скрытый слой
model.add(Dense(512, activation='relu',))
model.add(Dense(256, activation='relu',))
model.add(Dense(128, activation='relu',))
model.add(Dense(128, activation='relu',))
model.add(Dense(256, activation='relu',)) 
model.add(Dense(14, activation='softmax',)) # выходной слой
```

```python
model.summary()
```

```python
model.compile(
    loss='categorical_crossentropy',  # минимизируем кросс-энтропию
    optimizer='adam',  # так будет быстрее, позже узнаем что это :)
    metrics=['mse']  # выводим процент правильных ответов
)
```

```python
y_train_oh = keras.utils.to_categorical(Y_train, 14)
y_test_oh = keras.utils.to_categorical(Y_test, 14)

print(y_train_oh.shape)
print(y_train_oh[:5], Y_train[:5])
```

```python
history = model.fit(
                    X_train, 
                    y_train_oh,
                    batch_size=256,  # 64 объекта для подсчета градиента на каждом шаге
                    epochs=16,  # 10 проходов по датасету
                    validation_data=(X_test, y_test_oh)
                    )
```
